Data Description: Given a Bank customer, can we build a classifier that can determine whether they will leave
or not using Neural networks?

The dataset contains 10,000 sample points with 14 distinct features such as CustomerId, CreditScore, Geography, 
Gender, Age, Tenure, Balance etc. Know your data: https://www.kaggle.com/barelydedicated/bank-
customer-churn-modeling.

Businesses like banks which provide service have to worry about problem of 'Churn' i.e. customers leaving and
joining another service provider. It is important to understand which aspects of the service influence a
customer's decision in this regard. Management can concentrate efforts on improvement of service, keeping in 
mind these priorities.

The points distribution for this case is as follows:

# Read the data set,Drop the columns which are unique for all users like IDs (2.5 points)

```python
import numpy as np
import pandas as pd
bank_df=pd.read_csv("bank.csv")
print(bank_df.head())
print(" The column  names are features and the Target variable is Exited")
```

```python
#Checking for null values
bank_df.isnull().sum()
drop_list=["CustomerId","Exited","Surname"]
new_bank_df=bank_df.drop(drop_list,axis=1)
new_bank_df.head()
bank_df.head()
```

# Distinguish the feature and target set (2.5 points)

```python
print(" The column  names are features and the Target variable is Exited")
new_bank_df.groupby(["Geography"]).count()
```

# Encoding the Categorical Features Geography and Gender

```python
#Encoding all the object data types.
new_bank_obj = new_bank_df.select_dtypes(include=['object']).copy()
cleanup_nums = {"Gender": {"Female": 0, "Male": 1 },"Geography":{"France": 0, "Germany": 1,"Spain":2 }}
X_df=new_bank_df.replace(cleanup_nums, inplace=False)# Replacing the categorical Target variable with the binary encoded values.
X_df.head()
```

```python
X=X_df.values
#print("The following are the features\n",col_names)
Y_df=bank_df.iloc[:,13]
Y=Y_df.values
Y_df.head()
print(X_df.shape,Y_df.shape)
print("Target Set is \n",Y_df.head())
```

# Divide the data set into training and test sets ( 2.5 points)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=7)
```

```python
print(X_train)
```

# Normalize the train and test data (5 points)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Initialize & build the model (10 points)

```python
import tensorflow as tf
print(tf.__version__)
```

```python
tf.random.set_seed(42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Initialize Sequential model
model = Sequential()
```

```python
#Add input layer
model.add(Dense(11, input_dim = 11, kernel_initializer = 'uniform', activation='relu', name='input_layer'))
```

```python
#Add hidden layer
model.add(Dense(10, kernel_initializer = 'uniform', activation = 'relu', name='hidden_layer_1'))
```

```python
#Add Output layer
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid', name='output_layer'))
```

```python
#Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print('Neural Network Model Summary: ')
print(model.summary())
```

```python
# Fitting the ANN
model.fit(X_train,Y_train, epochs=100,validation_data=(X_test,Y_test))
```

# Predict the results using 0.5 as a threshold (5 points)

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred = (Y_pred > 0.5).astype(np.int)
bank_test_predict=pd.DataFrame(X_test,columns=X_df.columns)
bank_test_predict["Predicted"]=Y_pred
bank_test_predict["Actual"]=bank_df["Exited"]
bank_test_predict.head()
```

# Print the Accuracy score and confusion matrix (2.5 points)

```python
from sklearn import metrics
print(metrics.classification_report(Y_test, Y_pred))
```

```python
print("accuracy", metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
```
